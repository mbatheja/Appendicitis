```python
import streamlit_authenticator as stauth
#password list
passwords = ['121', '122', '123', '124', '125']
#Hashing the passwords
hashed_passwords = stauth.Hasher.hash_list(passwords)
#Print the hashed passwords
for password, hashed in zip(passwords, hashed_passwords):
    print(f"Password: {password}, Hashed Password: {hashed}")
```
