```python
import pandas as pd
data = pd.read_excel('app_data_his.xlsx')
data.head()
```

```python
#remove ultrasound data
data = data.iloc[:, :32]
```

```python
data.head()
```

```python
print(data.columns)
```

```python
data = data.drop(columns=["Segmented_Neutrophils", "Appendix_on_US", "Appendix_Diameter", "Length_of_Stay"])
```

```python
data = data.dropna()
```

```python
ketone_mapping = {"no": 0, "+": 1, "++": 1, "+++": 1}
data["Ketones_in_Urine"] = data["Ketones_in_Urine"].map(ketone_mapping)

wbc_mapping = {"no": 0, "+": 1, "++": 1, "+++": 1}
data["WBC_in_Urine"] = data["WBC_in_Urine"].map(ketone_mapping)

# numerical range provided for rbc in urine, so make levels 0-3
rbc_mapping = {"no": 0, "+": 1, "++": 2, "+++": 3}
data["RBC_in_Urine"] = data["RBC_in_Urine"].map(rbc_mapping)

# convert yes/no to 0/1 for neutrophilia
neutrophilia_mapping = {"no": 0, "yes": 1}
data["Neutrophilia"] = data["Neutrophilia"].map(neutrophilia_mapping)
```

```python
data.head()
```

```python
data.info()
```

```python
appendicitis_mapping = {"no appendicitis": 0, "appendicitis": 1}
data["Diagnosis"] = data["Diagnosis"].map(appendicitis_mapping)
```

```python
corr_matrix = data.corr()
```

```python
features = corr_matrix['Diagnosis'].abs().sort_values(ascending=False)
print("Selected Features:", features)
```
